## ATM INTERFACE
The atm interface project performs the below mentioned operations:-
 - Transactions History
 - Withdraw
 - Deposit
 - Transfer
 - Quit

```python
class ATM:
    def __init__(self, account_id, pin, initial_balance=0):
        self.account_id = 2021456987123
        self.pin = 1234
        self.balance = 1000

        # Simulate a transaction history
        self.transaction_history = []

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
            return

        self.balance -= amount
        self.transaction_history.append(("Withdrawal", amount))

    def deposit(self, amount):
        self.balance += amount
        self.transaction_history.append(("Deposit", amount))

    def transfer(self, to_account_id, amount):
        if amount > self.balance:
            print("Insufficient funds")
            return

        self.balance -= amount
        self.transaction_history.append(("Transfer", amount, to_account_id))

        # In a real system, you would need to update the balance of the receiving account

    def check_balance(self):
        return self.balance

    def view_transaction_history(self):
        for transaction in self.transaction_history:
            print(transaction)

def main():
    # Prompt the user to enter their account ID and PIN
    account_id = input("Enter your account ID: ")
    pin = input("Enter your PIN: ")

    # Create an instance of the ATM class
    atm = ATM(account_id, pin)

    while True:
        # Display the main menu
        print("\nMain Menu")
        print("1. Withdraw")
        print("2. Deposit")
        print("3. Transfer")
        print("4. Check balance")
        print("5. View transaction history")
        print("6. Quit")

        # Get the user's selection
        selection = int(input("\nEnter your selection: "))

        # Perform the selected operation
        if selection == 1:
            amount = float(input("Enter amount to withdraw: "))
            atm.withdraw(amount)
        elif selection == 2:
            amount = float(input("Enter amount to deposit: "))
            atm.deposit(amount)
        elif selection == 3:
            to_account_id = input("Enter recipient account ID: ")
            amount = float(input("Enter amount to transfer: "))
            atm.transfer(to_account_id, amount)
        elif selection == 4:
            balance = atm.check_balance()
            print("Your balance is:", balance)
        elif selection == 5:
            atm.view_transaction_history()
        elif selection == 6:
            break
        else:
            print("Invalid selection")

if __name__ == "__main__":
    main()
```
